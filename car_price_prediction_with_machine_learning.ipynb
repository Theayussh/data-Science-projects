Encoding the categorial data into numerical data.
petrol--->0
diesel--->1
CNG--->2

dealer--->0
individual-->1

manual--->0
automatic--->1

```python
car_data.replace({'Fuel_Type':{'Petrol':0, 'Diesel':1, 'CNG':2 }},inplace=True)
car_data.replace({'Transmission':{'Manual':0, 'Automatic':1}},inplace=True)
car_data.replace({'Selling_type':{'Dealer':0, 'Individual':1}},inplace=True)
```

```python
#again checking the distribution of data.
print(car_data.Fuel_Type.value_counts())
print(car_data.Selling_type.value_counts())
print(car_data.Transmission.value_counts())
```

Splitting the data to target data, trainging data and testing data


```python
X = car_data.drop(['Car_Name','Selling_Price'],axis=1)
Y = car_data['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train, Y_train)
```

```python
training_data_prediction = lr_model.predict(X_train)
```

```python
# R Square error
from sklearn import metrics
error_score = metrics.r2_score(Y_train, training_data_prediction)
print('R square score :', error_score)
```

Visualizing the actual prices and predicted prices

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price VS Predicted Price')
plt.show()
```

```python
test_data_prediction = lr_model.predict(X_test)
```

```python
error_score = metrics.r2_score(Y_test, test_data_prediction)
print('R square score :', error_score)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price VS Predicted Price')
plt.show()
```
